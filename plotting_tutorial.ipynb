# **Matplotlib**

## Scatterplot - basic
We start by defining some data. Let's suppose we collected age (in years), height (in cm), and weight (in kilograms) information from few patients visitng a clinic.

```python
people = ['Ann','Brandon','Chen','David','Emily','Farook',
          'Gagan','Hamish','Imran','Julio','Katherine','Lily']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
weight = [55,35,77,68,70,60,72,69,18,65,82,48]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
plt.scatter(age, height) # Ploting age on X-Axis and height on Y-Axis
plt.show() # The ploted image is stored at some place , this function will help to make the plot visualizable and presentable
```

Now let's modify the scatterplot a bit


*   Figure size
* Main title (and size the font)
* X- and Y-labels (and size the font)
* X- and Y-ticks (and size the font)
* Grid
* Text for labelling
* Legend
* Vertical (or horizontal) lines

You can read about all of them separately on the documentation of matplotlib website and can learn deeper about them.

**Link :-** https://matplotlib.org/stable/users/explain/quick_start.html


```python
# Set figure size
plt.figure(figsize=(10,6))

# Add a main title
plt.title("Plot of Age vs. Height (in cms)\n",fontsize=20, fontstyle='italic')

# X- and Y-label with fontsize
plt.xlabel("Age (years)",fontsize=16)
plt.ylabel("Height (cms)",fontsize=16)

# Turn on grid
plt.grid (True)

# Set Y-axis limit
plt.ylim(100,200)

# X- and Y-axis ticks customization with fontsize and placement
plt.xticks([i*5 for i in range(12)],fontsize=15)
plt.yticks(fontsize=15)

# Main plotting function with choice of color, marker size, and marker edge color
plt.scatter(x=age,y=height,c='green',s=100,edgecolors='k')

# Adding bit of text to the plot
plt.text(x=29,y=140,s="Height increaes up to around \n28 years and then tapers off",fontsize=10,
         rotation=0, linespacing=2)
plt.text(x=22,y=185,s="Nobody has a height beyond 180 cm",fontsize=13)

# Adding a vertical line
plt.vlines(x=28,ymin=100,ymax=180,linestyles='dashed',color='blue',lw=3)

# Adding a horizontal line
plt.hlines(y=180,xmin=0,xmax=55,linestyles='dashed',color='red',lw=3)

# Adding a legend
plt.legend(['Height in cms'],loc=2,fontsize=14)

# Final show method
plt.show()
```

## Bar Chart
We will now plot the same data on a barchart to understand well about the data.

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16, fontstyle='italic')
# Main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,color='green',edgecolor='k',alpha=0.8)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30) # we can use rotation to adjust the texts on the ticks.
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

# Histogram

```python
plt.figure(figsize=(7,5))
# Main plot function 'hist'
plt.hist(weight,color='blue',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

## Scatter PLot

```python
days = np.arange(1,31)
'''
When you call np.random.randn(n), it generates n random numbers from this standard normal distribution and returns them as an array.
These random numbers are typically used to introduce randomness or variability into simulations or to model uncertainty in statistical analyses.


The np.random.randn() function in NumPy generates an array of random numbers sampled from a standard normal distribution, also known as a Gaussian
distribution or bell curve. A standard normal distribution has a mean of 0 and a standard deviation of 1.
'''
candidate_A = 50+days*0.07+2*np.random.randn(30) # adds a linearly increasing number to the variable and adds randomly generated noise to it.
candidate_B = 50-days*0.1+3*np.random.randn(30)
```

```python
# Determine the minimum and maximum of stock prices
ymin = min(candidate_A.min(),candidate_B.min())
ymax = max(candidate_A.max(),candidate_B.max())
```

```python
# Set style
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12,5))
plt.title("Time series plot of poll percentage over a month\n",fontsize=20, fontstyle='italic')
plt.xlabel("Days",fontsize=16)
plt.ylabel("Poll percentage (%)",fontsize=16)
plt.grid (True)
plt.ylim(ymin*0.98,ymax*1.02)
plt.xticks([i*2 for i in range(16)],fontsize=14)
plt.yticks(fontsize=15)

# Main plotting function - plot (note markersize, lw (linewidth) arguments)
plt.plot(days,candidate_A,'o-',markersize=10,c='blue',lw=2)
plt.plot(days,candidate_B,'^-',markersize=10,c='green',lw=2)

plt.legend(['Poll percentage of candidate A (%)', 'Poll percentage of candidate B (%)'],loc=2,fontsize=14)
plt.show()
```

## Boxplot

image.png

```python
plt.style.use('ggplot')
# Note how to convert default numerical x-axis ticks to the list of string by passing two lists
plt.boxplot(x=[candidate_A,candidate_B],showmeans=True)
plt.grid(True)
plt.xticks([1,2],['Candidate A','Candidate B'])
plt.yticks(fontsize=10)
plt.show()
```
